**Problem Statement**
Online book retailers and publishing houses face a critical challenge in predicting a book's commercial success before its release or during its early market presence. With thousands of new titles published annually, making accurate revenue forecasts and inventory decisions can mean the difference between profitable operations and costly overstock or stockouts. This project addresses this business need by framing the problem as a regression task to predict a book's popularity and market performance, using "reviews count" as a proxy for commercial success and reader engagement from best selling online books data.

Success will be measured by RMSE (Root Mean Square Error) as the primary metric, providing interpretable error margins in terms of review counts, with R² serving as a secondary metric to explain the variance captured by our model. From a business perspective, a successful model would enable publishers to identify potential bestsellers earlier in their lifecycle, optimize initial print runs, and make data-driven decisions about marketing investments. The business KPI would be a 15% improvement in inventory turnover rate, a 10% reduction in unsold inventory and 10% improvement in marketing ROI, achieved through more accurate sales forecasting based on predicted review volumes.

```python
# ========== 6) CRITICAL THINKING → ETHICAL AI & BIAS AUDITING ==========
print("\n" + "="*60)
print("ETHICAL AI & BIAS AUDITING")
print("="*60)
print("Ensuring the model is fair, explainable, and responsible")
print("="*60)

# 🟢 FIXED: Check if df_feat exists, if not use df
if 'df_feat' not in dir():
    df_feat = df.copy()
    print("Using original dataframe for bias analysis")

# 🟢 FIXED: Use the correct feature set from Model Implementation
if 'final_features' in dir():
    bias_features = final_features
    print(f"✅ Using feature set from Model Implementation: {len(bias_features)} features")
else:
    # Fallback to selected features
    if 'final_selected_features' in dir():
        bias_features = final_selected_features
        print(f"⚠️ Using fallback feature set: {len(bias_features)} features")
    else:
        bias_features = [col for col in df_feat.select_dtypes(include=[np.number]).columns
                         if col not in ['id', 'reviews count', 'log_reviews']]
        print(f"⚠️ Using all numeric features: {len(bias_features)} features")

# ========== 6a) MODEL EXPLAINABILITY WITH SHAP ==========
print("\n" + "-"*40)
print("6a) MODEL EXPLAINABILITY - SHAP ANALYSIS")
print("-"*40)
print("Using SHAP to answer: 'Why did the model make this prediction?'")

try:
    import shap

    # 🟢 FIXED: Get the best model for explanation from Model Implementation
    print("\nGenerating SHAP explanations for the best model...")

    # Identify best model from results_df
    if 'results_df' in dir() and len(results_df) > 0:
        best_model_name = results_df.iloc[0]['Model']
        base_name = best_model_name.replace(' (Tuned)', '')
        print(f"Best model: {best_model_name}")
    else:
        best_model_name = "Decision Tree"
        base_name = "Decision Tree"
        print("Using Decision Tree as default for explanation")

    # 🟢 FIXED: Get the model instance from Model Implementation
    model_for_explanation = None

    # Check for the final model from Model Implementation
    if 'final_model' in dir() and final_model is not None:
        model_for_explanation = final_model
        print("   Using final_model from Model Implementation")
    elif 'best_model_instance' in dir() and best_model_instance is not None:
        model_for_explanation = best_model_instance
        print("   Using best_model_instance from Model Implementation")
    elif 'models' in dir() and base_name in models:
        model_for_explanation = models[base_name]
        print("   Using model from models dictionary")
    else:
        # Fallback to Random Forest
        from sklearn.ensemble import RandomForestRegressor
        model_for_explanation = RandomForestRegressor(n_estimators=100, random_state=42)
        if 'X_train' in dir() and 'y_train' in dir():
            model_for_explanation.fit(X_train, y_train)
            print("   - Trained Random Forest as fallback for explanation")
        else:
            print("   - Cannot train model - using pre-trained model from earlier section")

    # 🟢 FIXED: Use bias_features for SHAP (consistent with Model Implementation)
    if model_for_explanation is not None and hasattr(model_for_explanation, 'feature_importances_'):
        # Create explainer
        explainer = shap.TreeExplainer(model_for_explanation)

        # 🟢 FIXED: Use the correct feature matrix
        if 'X_test' in dir():
            # Get the feature matrix for test data using bias_features
            X_test_bias = df_feat[bias_features].fillna(0).loc[X_test.index] if hasattr(X_test, 'index') else None

            if X_test_bias is not None and len(X_test_bias) > 0:
                if len(X_test_bias) > 100:
                    X_test_sample = X_test_bias.sample(100, random_state=42)
                    shap_values = explainer.shap_values(X_test_sample)
                    X_display = X_test_sample
                else:
                    shap_values = explainer.shap_values(X_test_bias)
                    X_display = X_test_bias

                # Summary plot
                plt.figure(figsize=(12, 8))
                shap.summary_plot(shap_values, X_display, feature_names=bias_features, show=False)
                plt.title(f'SHAP Summary Plot - {best_model_name}', fontsize=14)
                plt.tight_layout()
                plt.show()

                # Feature importance bar plot
                plt.figure(figsize=(10, 6))
                shap.summary_plot(shap_values, X_display, feature_names=bias_features,
                                  plot_type="bar", show=False)
                plt.title(f'SHAP Feature Importance - {best_model_name}', fontsize=14)
                plt.tight_layout()
                plt.show()

                # Individual prediction explanation
                print("\nExplaining individual predictions:")
                for i in range(min(3, len(X_display))):
                    print(f"\n   Sample {i+1}:")
                    if 'y_test' in dir():
                        actual_val = y_test.iloc[i] if hasattr(y_test, 'iloc') else y_test[i]
                        print(f"   Actual: {actual_val:.0f} reviews")

                    pred_val = model_for_explanation.predict([X_display.iloc[i].values])[0]
                    print(f"   Predicted: {pred_val:.0f} reviews")

                    # Waterfall plot
                    shap.waterfall_plot(shap.Explanation(values=shap_values[i],
                                                         base_values=explainer.expected_value,
                                                         data=X_display.iloc[i].values,
                                                         feature_names=bias_features),
                                       show=False)
                    plt.title(f'SHAP Waterfall Plot - Sample {i+1}')
                    plt.tight_layout()
                    plt.show()
            else:
                print("   - Test data not available for SHAP analysis")
        else:
            print("   - X_test not available for SHAP analysis")
    else:
        print("   - Model not tree-based or not available, skipping detailed SHAP analysis")

except ImportError:
    print("\n⚠ SHAP library not installed. Installing now...")
    !pip install shap
    print("Please re-run this section after installation.")
except Exception as e:
    print(f"\n⚠ Error in SHAP analysis: {e}")
    print("   Continuing with bias detection...")

# ========== 6b) BIAS DETECTION QUESTIONS ==========
print("\n" + "-"*40)
print("6b) BIAS DETECTION ANALYSIS")
print("-"*40)
print("Answering critical fairness questions:")
print("-"*20)

# 🟢 FIXED: Create a copy for bias analysis using the correct feature set
df_bias = df_feat.copy()

# Identify potential protected attributes and their proxies
print("\n1. IDENTIFYING PROTECTED ATTRIBUTES AND PROXIES:")
print("-"*40)

protected_attributes = {
    'Genre': ['genre'],
    'Book Format': ['form'],
    'Price Accessibility': ['price'],
    'Author': ['author']
}

print("Potential protected attributes in book recommendation systems:")
for attr, features in protected_attributes.items():
    print(f"\n   • {attr}:")
    for feat in features:
        if feat in df_bias.columns:
            if df_bias[feat].dtype == 'object' or df_bias[feat].nunique() < 20:
                unique_count = df_bias[feat].nunique()
                print(f"     - {feat} ({unique_count} categories)")
                value_counts = df_bias[feat].value_counts().head(5)
                print(f"       Top categories: {dict(value_counts)}")
            else:
                print(f"     - {feat} (numerical, range: {df_bias[feat].min():.2f} to {df_bias[feat].max():.2f})")

# 🟢 FIXED: Use bias_features for proxy detection
print("\n   Potential Proxy Features:")
proxy_features = []

if 'bias_features' in dir() and len(bias_features) > 0:
    for col in bias_features:
        # Check correlation with author
        if 'author' in df_bias.columns and col in df_bias.columns and col != 'author':
            if df_bias[col].dtype in ['int64', 'float64']:
                try:
                    author_encoded = pd.factorize(df_bias['author'])[0]
                    corr = abs(df_bias[col].corr(pd.Series(author_encoded, index=df_bias.index)))
                    if corr > 0.3 and not pd.isna(corr):
                        proxy_features.append((col, 'author', corr))
                except:
                    pass

        # Check correlation with genre
        if 'genre' in df_bias.columns and col in df_bias.columns and col != 'genre':
            if df_bias[col].dtype in ['int64', 'float64']:
                try:
                    genre_encoded = pd.factorize(df_bias['genre'])[0]
                    corr = abs(df_bias[col].corr(pd.Series(genre_encoded, index=df_bias.index)))
                    if corr > 0.3 and not pd.isna(corr):
                        proxy_features.append((col, 'genre', corr))
                except:
                    pass

if proxy_features:
    print(f"     Found {len(proxy_features)} potential proxy features:")
    for feat, proxy_for, corr in proxy_features[:5]:
        print(f"     - '{feat}' correlated with {proxy_for} (r={corr:.2f}) - potential proxy")
else:
    print("     - No strong proxy features detected")

print("\n2. PERFORMANCE ACROSS GROUPS:")
print("-"*40)
print("Evaluating if model performs equally well across groups...")

# 🟢 FIXED: Get predictions from Model Implementation
y_pred_test = None

if 'y_pred_tuned' in dir():
    y_pred_test = y_pred_tuned
    print("Using tuned model predictions")
elif 'predictions' in dir() and len(predictions) > 0:
    if 'best_final_model_name' in dir():
        best_model_key = best_final_model_name.replace(' (Tuned)', '')
        if best_model_key in predictions:
            y_pred_test = predictions[best_model_key]
            print(f"Using {best_model_key} predictions")
    else:
        y_pred_test = list(predictions.values())[0]
        print("Using first available model predictions")
elif 'y_pred_best' in dir():
    y_pred_test = y_pred_best
    print("Using y_pred_best")
elif 'final_model' in dir() and 'X_test' in dir():
    y_pred_test = final_model.predict(X_test)
    print("Using final_model predictions")

# 🟢 FIXED: Use bias_features for test data
if y_pred_test is not None and 'X_test' in dir() and 'y_test' in dir():
    # Create test dataframe with indices
    if hasattr(X_test, 'index'):
        test_indices = X_test.index
    else:
        test_indices = range(len(y_pred_test))

    # Ensure y_test is properly formatted
    if hasattr(y_test, 'values'):
        y_test_values = y_test.values
    else:
        y_test_values = y_test

    df_test_results = pd.DataFrame({
        'actual': y_test_values[:len(y_pred_test)],
        'predicted': y_pred_test,
        'error': abs(y_test_values[:len(y_pred_test)] - y_pred_test)
    }, index=test_indices[:len(y_pred_test)])

    # Merge with original features
    for col in df_bias.columns:
        if col in df_bias.columns:
            try:
                df_test_results[col] = df_bias.loc[test_indices[:len(y_pred_test)], col].values
            except:
                pass

    # Analyze performance across groups
    groups_to_analyze = {'Genre': 'genre', 'Format': 'form'}
    fairness_metrics = []

    for group_name, group_col in groups_to_analyze.items():
        if group_col in df_test_results.columns:
            print(f"\n   Analyzing by {group_name}:")

            group_metrics = df_test_results.groupby(group_col).agg({
                'error': ['mean', 'std', 'count'],
                'actual': 'mean',
                'predicted': 'mean'
            }).round(2)

            group_metrics.columns = ['MAE', 'Error Std', 'Count', 'Avg Actual', 'Avg Predicted']

            overall_median = df_test_results['predicted'].median()
            for group in group_metrics.index:
                group_data = df_test_results[df_test_results[group_col] == group]
                demo_parity = (group_data['predicted'] > overall_median).mean()
                group_metrics.loc[group, 'Demographic Parity'] = demo_parity

            print(group_metrics)

            mae_diff = group_metrics['MAE'].max() - group_metrics['MAE'].min()
            parity_diff = group_metrics['Demographic Parity'].max() - group_metrics['Demographic Parity'].min()

            fairness_metrics.append({
                'Group': group_name,
                'Max MAE Difference': mae_diff,
                'Max Parity Difference': parity_diff,
                'Group Count': len(group_metrics)
            })

    # Summary of fairness metrics
    print("\n   FAIRNESS METRICS SUMMARY:")
    for metric in fairness_metrics:
        print(f"\n   • {metric['Group']} ({metric['Group Count']} groups):")
        print(f"     - Equalized Odds check: MAE difference = {metric['Max MAE Difference']:.2f}")
        if metric['Max MAE Difference'] < 5000:
            print("       ✓ Model accuracy is relatively consistent across groups")
        else:
            print("       ⚠ Model accuracy varies significantly across groups")

        print(f"     - Demographic Parity difference = {metric['Max Parity Difference']:.3f}")
        if metric['Max Parity Difference'] < 0.15:
            print("       ✓ Prediction rates are similar across groups")
        else:
            print("       ⚠ Prediction rates differ across groups")
else:
    print("   Unable to evaluate group performance - predictions not available")

print("\n3. HARM ANALYSIS - Would different groups be harmed by model errors?")
print("-"*40)

# 🟢 FIXED: Analyze error patterns using bias_features
if 'df_test_results' in dir():
    df_test_results['error_type'] = 'Correct (±25%)'
    df_test_results.loc[df_test_results['predicted'] > df_test_results['actual'] * 1.5, 'error_type'] = 'Severe Over-prediction (>50% high)'
    df_test_results.loc[df_test_results['predicted'] < df_test_results['actual'] * 0.5, 'error_type'] = 'Severe Under-prediction (>50% low)'

    groups_to_analyze = {'Genre': 'genre', 'Format': 'form'}

    for group_name, group_col in groups_to_analyze.items():
        if group_col in df_test_results.columns:
            print(f"\n   Error distribution by {group_name}:")
            error_by_group = pd.crosstab(
                df_test_results[group_col],
                df_test_results['error_type'],
                normalize='index'
            ) * 100

            group_counts = df_test_results[group_col].value_counts()
            significant_groups = group_counts[group_counts >= 5].index

            if len(significant_groups) > 0:
                display_data = error_by_group.loc[significant_groups]
                print(display_data.round(1))
            else:
                print("     Insufficient data per group for meaningful analysis")
else:
    print("   Unable to analyze error patterns - test results not available")

# ========== 6c) LIMITATIONS DOCUMENTATION ==========
print("\n" + "-"*40)
print("6c) MODEL LIMITATIONS")
print("-"*40)
print("Honest documentation of model constraints:")
print("-"*20)

limitations = []

# 1. Class/Data imbalance
print("\n1. CLASS/DATA IMBALANCE:")
if 'reviews count' in df.columns:
    review_skew = df['reviews count'].skew()
    review_mean = df['reviews count'].mean()
    review_median = df['reviews count'].median()

    print(f"   • Target variable statistics:")
    print(f"     - Mean: {review_mean:.0f}, Median: {review_median:.0f}")
    print(f"     - Skewness = {review_skew:.2f}")

    if abs(review_skew) > 1:
        print(f"   • Target variable is highly skewed")
        print(f"   • Addressed using: Log transformation (log_reviews feature)")
        limitations.append("Severe class imbalance in review counts")
    else:
        print(f"   • Target variable has moderate skew")

# 2. Feature set used
print("\n2. FEATURE SET USED:")
if 'bias_features' in dir():
    print(f"   • Number of features used: {len(bias_features)}")
    essential = ['format_Paperback', 'format_Hardcover', 'price_cat_Standard']
    present = [f for f in essential if f in bias_features]
    print(f"   • Essential format/price features included: {len(present)}/3")
    if len(present) == 3:
        print("     ✓ All essential features included")
    else:
        print(f"     ⚠ Missing: {[f for f in essential if f not in bias_features]}")

# 3. Generalization limitations
print("\n3. GENERALIZATION LIMITATIONS:")
if 'X_train' in dir():
    print(f"   • Training data size: {len(X_train)} samples")
if 'X_test' in dir():
    print(f"   • Test data size: {len(X_test)} samples")
if 'best_final_r2' in dir():
    print(f"   • Model R² on test set: {best_final_r2:.3f}")
    if best_final_r2 > 0.5:
        print("   ✓ Good explanatory power - likely to generalize well")
    else:
        print("   ⚠ Limited explanatory power - may not generalize well")

# 4. Domain-specific limitations
print("\n4. DOMAIN-SPECIFIC LIMITATIONS:")
domain_limitations = [
    "• Model trained only on bestselling books from 2023 - may not represent all books",
    "• Review count as proxy for popularity ignores qualitative aspects (sentiment, content)",
    "• Temporal bias - recent books may have accumulated fewer reviews",
    "• Platform bias - Amazon reviews may not represent all reader demographics",
    "• Limited sample size (100 books) - may not capture full market diversity"
]

for limitation in domain_limitations:
    print(f"   {limitation}")
    limitations.append(limitation)

# 5. Feature limitations
print("\n5. FEATURE LIMITATIONS:")
print("   Missing features that could improve predictions:")
missing_features = [
    "Marketing budget and promotional activity",
    "Social media sentiment and buzz (Twitter, TikTok, Instagram)",
    "Pre-order numbers and early sales data",
    "Author's social media following and engagement",
    "Competing book releases in same genre",
    "Seasonal factors (holiday shopping, summer reading)",
    "Awards and nominations",
    "Book club selections (Oprah, Reese, etc.)"
]

for feature in missing_features[:8]:
    print(f"     - {feature}")

# ========== SUMMARY: BIAS AUDIT REPORT ==========
print("\n" + "="*60)
print("BIAS AUDIT SUMMARY REPORT")
print("="*60)

print("\nFAIRNESS METRICS CHECK:")
print("-"*60)
print(f"{'Metric':<25} {'Status':<10} {'Value':<15} {'Interpretation'}")
print("-"*60)

if 'fairness_metrics' in dir() and fairness_metrics:
    for metric in fairness_metrics:
        if metric['Max MAE Difference'] < 5000:
            status = "✓ PASS"
        else:
            status = "⚠ REVIEW"
        print(f"{'Equalized Odds (' + metric['Group'] + ')':<25} {status:<10} {f'Diff: {metric["Max MAE Difference"]:.0f}':<15} | Accuracy varies across groups")

        if metric['Max Parity Difference'] < 0.15:
            status = "✓ PASS"
        else:
            status = "⚠ REVIEW"
        print(f"{'Demographic Parity (' + metric['Group'] + ')':<25} {status:<10} {f'Diff: {metric["Max Parity Difference"]:.3f}':<15} | Prediction rates vary")
else:
    print("Unable to compute fairness metrics - insufficient data")

print("\nIDENTIFIED LIMITATIONS:")
print("-"*60)
unique_limitations = list(set(limitations))[:10]
for i, limitation in enumerate(unique_limitations, 1):
    print(f"{i}. {limitation}")

print("\nRECOMMENDATIONS FOR MITIGATION:")
print("-"*60)
print("""
1. Data Collection Improvements:
   • Gather more diverse book data beyond bestsellers (include mid-list and niche titles)
   • Include demographic information about authors and target audiences
   • Collect data from multiple platforms (Goodreads, BookBub, Library systems)
   • Add temporal data to track review accumulation over time

2. Model Enhancements:
   • Implement fairness constraints during training (reweighting, adversarial debiasing)
   • Use ensemble methods with different feature sets to reduce bias
   • Regular retraining with updated data (quarterly)
   • Consider multi-task learning to predict both review count and rating

3. Deployment Safeguards:
   • Monitor predictions across different book categories and formats
   • Implement human review for books with high prediction uncertainty
   • Provide explainability interfaces for stakeholders
   • Regular bias audits (quarterly) with updated fairness metrics

4. Transparency Measures:
   • Document known limitations in model cards
   • Provide confidence intervals with all predictions
   • Allow users to provide feedback on recommendations
   • Publish fairness metrics alongside model performance

5. Business Process Integration:
   • Use model as one input in a broader decision framework
   • Combine with human expertise for final decisions
   • A/B test model recommendations against current processes
   • Establish clear escalation paths for edge cases
""")

print("\n" + "="*60)
print("ETHICAL AI ANALYSIS COMPLETE")
print("="*60)
print("Model is ready for responsible deployment with ongoing monitoring")
```
